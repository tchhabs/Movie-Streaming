```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt

# File to Load
file_to_load = "./Resources/MoviesOnStreamingPlatforms_updated.csv"

# Read Purchasing File and store into Pandas data frame
csv = pd.read_csv(file_to_load, engine='python')
csv.head()
```

```python
moviecount = (len(csv))
moviecount
```

```python
newdata = csv[(csv['Year'] > 2009) & (csv['Year'] < 2019)]
newdata
```

## Which streaming platform has the most movies?

```python
total_movies=newdata["Title"].count()
total_movies
```

```python
Netflix = newdata["Netflix"].value_counts()
Netflix[1]
```

```python
Hulu = newdata["Hulu"].value_counts()
Hulu[1]
```

```python
PrimeVideo = newdata["Prime Video"].value_counts()
PrimeVideo[1]
```

```python
Disney = newdata["Disney+"].value_counts()
Disney[1]
```

```python
number_movies= pd.DataFrame({"Number of Netflix Movies": [Netflix[1]], "Number of Hulu Movies": [Hulu[1]], 
                             "Number of Prime Video Movies":[PrimeVideo[1]], "Number of Disney+ Movies":[Disney[1]]})
number_movies
```

```python
number_movies.plot(kind = "bar", figsize = (10,5))

plt.title("Number of Movies on Each Streaming Platform")
plt.xlabel("Streaming Platforms")
plt.ylabel("Number of Movies")

plt.savefig("figure1.png")
plt.show()
plt.tight_layout
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Netflix', 'Hulu', 'Prime Video', 'Disney+'
sizes = [2440, 590, 6156, 135]
explode = (0, 0, 0, 0)  

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal') 
plt.savefig("figure2.png")
plt.show()
```

## Netflix: Rotten Tomatoes

```python
RTrating = newdata[('Rotten Tomatoes')].sort_values()
RT2 = RTrating.dropna(how = 'any')
RTdf = pd.DataFrame(newdata[['Netflix','Rotten Tomatoes', 'Title']])
RTdf.dropna(inplace=True)
RTdf = RTdf[RTdf.Netflix != 0]
RTdf
```

```python
RTdf.to_csv("Resources/netflix.csv")
netflix = pd.read_csv("Resources/netflix.csv")
netflix["Rotten Tomatoes"] = netflix["Rotten Tomatoes"].str.replace('%', "")
netflix["Rotten Tomatoes"] = pd.to_numeric(netflix["Rotten Tomatoes"])
netflix["Rotten Tomatoes"].dtypes
```

```python
RTrating = netflix[('Rotten Tomatoes')].sort_values()
RT2 = RTrating.dropna(how = 'any')
RTdf = pd.DataFrame(netflix[['Netflix','Rotten Tomatoes', 'Title']])
RTdf.dropna(inplace=True)
RTdf = RTdf[RTdf.Netflix != 0]
RTdf
```

```python
# Make bins based on Percentages
bins = [0, 20, 40, 60,80, 100]
percentages = ["Terrible", "Fair", "Good", "Great", "Awesome"]
netflix["Rotten Tomatoes Label"] = pd.cut(netflix["Rotten Tomatoes"], bins, labels= percentages, include_lowest=True)

netflix.head()
```

```python
netflixbar = netflix.groupby("Rotten Tomatoes Label")
netflixbar2 = netflixbar["Netflix"].count()
```

```python
netflixbar2.plot.bar(color = "red", edgecolor = "darkred", linewidth = "2")
plt.title("Netflix: Rotten Tomatoes Ratings")
plt.xlabel("Rotten Tomatoes Ratings")
plt.ylabel("Number of Movies")
plt.savefig("figure3.png")
```
